```python
import pygame
import time
import random

pygame.init()

white = (255,255,255)
black = (0,0,0)
red = (255,0,0)
green = (0,155,0)
yellow= (255,255,0)

display_width = 600
display_height  = 480

gameDisplay = pygame.display.set_mode((display_width,display_height))
pygame.display.set_caption('貪食蛇')



clock = pygame.time.Clock()

block_size = 10

FPS = 20

font = pygame.font.SysFont("comicsansms", 25)

def snake(block_size, snakelist):
    for XnY in snakelist:
        pygame.draw.rect(gameDisplay, yellow, [XnY[0],XnY[1],block_size,block_size])
    

def message_to_screen(msg,color):
    screen_text = font.render(msg, True, color)
    gameDisplay.blit(screen_text, [display_width/4, display_height/4])



def gameLoop():
    gameExit = False
    gameOver = False
    k=False
    lead_x = display_width/2
    lead_y = display_height/2
    total = 0
    lead_x_change = 0
    lead_y_change = 0
    tt = False
    
    
    

    snakeList = []
    snakeLength = 1
    randAppleXDDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
    randAppleYDDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
    randAppleYDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
    randAppleXDD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
    randAppleYD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
    randAppleXD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
    randAppleX = round(random.randrange(0, display_width-block_size)/10.0)*10.0
    randAppleY = round(random.randrange(0, display_height-block_size)/10.0)*10.0
    randAppleYDDDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
    randAppleXDDDD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
    randAppleYDDDDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
    randAppleXDDDDD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
    while randAppleXD != randAppleX and randAppleXD != randAppleX:
        randAppleYD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
        randAppleXD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
    while randAppleXDD != randAppleX and randAppleXD != randAppleX:
        randAppleYDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
        randAppleXDD = round(random.randrange(0, display_width-block_size)/10.0)*10.0    
    while randAppleXDD != randAppleX and randAppleXD != randAppleX:
        randAppleYDDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
        randAppleXDDD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
    while randAppleXDDD != randAppleX and randAppleXD != randAppleX:
        randAppleYDDDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
        randAppleXDDDD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
    while randAppleXDDDD != randAppleX and randAppleXD != randAppleX:
        randAppleYDDDDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
        randAppleXDDDDD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
    background = pygame.image.load("C:\Users\user\Desktop\images.jpg")

    while not gameExit:
        
        while gameOver == True:
            gameDisplay.fill(0)
            gameDisplay.blit(background, (0, 0))
            message_to_screen("Press R to restart.Press EXC to quit.", red)
            pygame.display.update()

            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_ESCAPE:
                        gameExit = True
                        gameOver = False
                    if event.key == pygame.K_r:
                        gameLoop()

        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                gameExit = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    if lead_x_change>0:
                        lead_x_change = lead_x_change
                        lead_y_change = 0
                    else:
                        lead_x_change = -block_size
                        lead_y_change = 0
                elif event.key == pygame.K_RIGHT:
                    if lead_x_change<0:
                        lead_x_change = lead_x_change
                        lead_y_change = 0
                    else:
                        lead_x_change = block_size
                        lead_y_change = 0
                elif event.key == pygame.K_UP:
                    if lead_y_change>0:
                        lead_y_change = lead_y_change
                        lead_x_change = 0
                    else:
                        lead_y_change = -block_size
                        lead_x_change = 0
                elif event.key == pygame.K_DOWN:
                    if lead_y_change<0:
                        lead_y_change = lead_y_change
                        lead_x_change = 0
                    else:
                        lead_y_change = block_size
                        lead_x_change = 0
            

        if lead_x >= display_width:
            gameOver = True
        if lead_x < 0:
            gameOver = True
        if lead_y >= display_height:
            gameOver = True
        if  lead_y < 0:
            gameOver = True
            
        lead_x += lead_x_change
        lead_y += lead_y_change
        
        gameDisplay.fill(0)
        gameDisplay.blit(background, (0, 0))
        while k == True:
            if randAppleX == randAppleXD and randAppleY == randAppleYD:
                k = False
                randAppleX = round(random.randrange(0, display_width-block_size)/10.0)*10.0
            else:
                k = True
                break
        pygame.draw.rect(gameDisplay, red, [randAppleX, randAppleY, block_size,block_size])
        pygame.draw.rect(gameDisplay, white, [randAppleXD, randAppleYD, block_size,block_size])
        pygame.draw.rect(gameDisplay, white, [randAppleXDD, randAppleYDD, block_size,block_size])
        pygame.draw.rect(gameDisplay, white, [randAppleXDDD, randAppleYDDD, block_size,block_size])
        pygame.draw.rect(gameDisplay, white, [randAppleXDDDD, randAppleYDDDD, block_size,block_size])
        pygame.draw.rect(gameDisplay, white, [randAppleXDDDDD, randAppleYDDDDD, block_size,block_size])
        snakeHead = []
        snakeHead.append(lead_x)
        snakeHead.append(lead_y)
        snakeList.append(snakeHead)

        if len(snakeList) > snakeLength:
            del snakeList[0]

        for eachSegment in snakeList[:-1]:
            if eachSegment == snakeHead:
                gameOver = True

        
        snake(block_size, snakeList)

        
        pygame.display.update()
        
        if lead_x == randAppleX and lead_y == randAppleY:
            randAppleX = round(random.randrange(0, display_width-block_size)/10.0)*10.0
            randAppleY = round(random.randrange(0, display_height-block_size)/10.0)*10.0
            snakeLength += 1
            randAppleXD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
            randAppleYD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
            randAppleXDD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
            randAppleYDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
            randAppleXDDD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
            randAppleYDDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
            randAppleXDDDD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
            randAppleYDDDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
            randAppleXDDDDD = round(random.randrange(0, display_width-block_size)/10.0)*10.0
            randAppleYDDDDD = round(random.randrange(0, display_height-block_size)/10.0)*10.0
        if lead_x == randAppleXD and lead_y == randAppleYD:
            gameOver = True
        if lead_x == randAppleXDD and lead_y == randAppleYDD:
            gameOver = True
        if lead_x == randAppleXDDD and lead_y == randAppleYDDD:
            gameOver = True
        if lead_x == randAppleXDDDD and lead_y == randAppleYDDDD:
            gameOver = True
        if lead_x == randAppleXDDDDD and lead_y == randAppleYDDDDD:
            gameOver = True
        clock.tick(FPS)
        
    pygame.quit()
    quit()


gameLoop()
```
